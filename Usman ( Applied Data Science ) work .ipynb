# Import necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the dataset

```python
data = pd.read_csv("COVID-19 Coronavirus.csv")
```

```python
data
```

# Data cleaning

```python
data_cleaned = data.dropna()  # Drop rows with missing values
data_filled = data.fillna(0)  # Fill missing values with zeros
```

# Visualizations

```python
# Histogram for Total Cases
plt.figure(figsize=(8, 6))
sns.histplot(data_cleaned['Total Cases'], bins=20, kde=True)
plt.title('Histogram of Total Cases')
plt.xlabel('Total Cases')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter Plot for Total Deaths vs Total Cases
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Total Cases', y='Total Deaths', data=data_cleaned)
plt.title('Scatter Plot of Total Deaths vs Total Cases')
plt.xlabel('Total Cases')
plt.ylabel('Total Deaths')
plt.show()
```

```python
# Select numeric columns for correlation matrix
numeric_cols = data_filled.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = data_filled[numeric_cols].corr()

# Heatmap for Correlation Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Box Plot for Death Percentage by Continent
plt.figure(figsize=(12, 8))
sns.boxplot(x='Continent', y='Death percentage', data=data_cleaned)
plt.title('Box Plot of Death Percentage by Continent')
plt.xlabel('Continent')
plt.ylabel('Death Percentage')
plt.show()
```
